# CAMADA SILVER

```python
from pyspark.sql import functions as F

# ORDERS 

# 1. LER DA BRONZE
df_bronze_orders = spark.table("olist.Bronze.bronze_orders")

# 2. APLICAR TRANSFORMAÇÕES  (Limpeza)
df_silver_orders = df_bronze_orders.withColumn(
    # Converte a coluna de status para maiúsculas (Padronização)
    "order_status", F.upper(F.col("order_status"))
) \
.withColumn(
    # Converte a string de data para um formato timestamp (Tipagem)
    "order_purchase_timestamp", F.to_timestamp(F.col("order_purchase_timestamp"))
) \
.withColumn(
    "order_approved_at", F.to_timestamp(F.col("order_approved_at"))
) \
.withColumn(
    "order_delivered_carrier_date", F.to_timestamp(F.col("order_delivered_carrier_date"))
) \
.withColumn(
    "order_delivered_customer_date", F.to_timestamp(F.col("order_delivered_customer_date"))
) \
.withColumn(
    "order_estimated_delivery_date", F.to_timestamp(F.col("order_estimated_delivery_date"))
)

# 3. SALVAR NA SILVER
nome_tabela_silver = "olist.Silver.silver_orders"
df_silver_orders.write.format("delta") \
                .mode("overwrite") \
                .saveAsTable(nome_tabela_silver)

print(f"Tabela {nome_tabela_silver} criada com sucesso!")
```

```python
# CUSTOMERS

# 1. LER DA BRONZE
df_bronze_customers = spark.table("olist.Bronze.bronze_customers")

# 2. APLICAR TRANSFORMAÇÕES
df_silver_customers = df_bronze_customers.withColumn(
    "customer_state", F.upper(F.col("customer_state"))
) \
.withColumn(
    "customer_city", F.upper(F.col("customer_city"))
)

# 3. SALVAR NA SILVER
nome_tabela_silver = "olist.Silver.silver_customers"
df_silver_customers.write.format("delta") \
                   .mode("overwrite") \
                   .saveAsTable(nome_tabela_silver)

print(f"Tabela {nome_tabela_silver} criada com sucesso!")
```

```python
# ORDER_ITEMS

# 1. LER DA BRONZE
df_bronze_items = spark.table("olist.Bronze.bronze_order_items") # Atenção ao nome da tabela

# 2. APLICAR TRANSFORMAÇÕES
df_silver_items = df_bronze_items.withColumn(
    # Convertendo colunas de valor para tipo numérico (Double)
    "price", F.col("price").cast("double")
) \
.withColumn(
    "freight_value", F.col("freight_value").cast("double")
) \
.withColumn(
    "shipping_limit_date", F.to_timestamp(F.col("shipping_limit_date"))
)


# 3. SALVAR NA SILVER
nome_tabela_silver = "olist.Silver.silver_order_items"
df_silver_items.write.format("delta") \
               .mode("overwrite") \
               .saveAsTable(nome_tabela_silver)

print(f"Tabela {nome_tabela_silver} criada com sucesso!")
```

```python
# PRODUCTS

df_bronze_products = spark.table("olist.Bronze.bronze_products")
df_bronze_translation = spark.table("olist.Bronze.bronze_product_category_name_translation")

# JUNTAR AS DUAS TABELAS 

df_joined = df_bronze_products.join(
    df_bronze_translation,
    ["product_category_name"], 
    "left"
)

# 2. APLICAR TRANSFORMAÇÕES 

df_silver_products = df_joined.withColumn(
    "product_category_name", F.upper(F.col("product_category_name"))
) \
.withColumn(
    "product_category_name_english", F.upper(F.col("product_category_name_english"))
) \
.withColumn(
    "product_name_lenght", F.col("product_name_lenght").cast("double")
) \
.withColumn(
    "product_description_lenght", F.col("product_description_lenght").cast("double")
) \
.withColumn(
    "product_photos_qty", F.col("product_photos_qty").cast("double")
) \
.withColumn(
    "product_weight_g", F.col("product_weight_g").cast("double")
) \
.withColumn(
    "product_length_cm", F.col("product_length_cm").cast("double")
) \
.withColumn(
    "product_height_cm", F.col("product_height_cm").cast("double")
) \
.withColumn(
    "product_width_cm", F.col("product_width_cm").cast("double")
)

# 3. SALVAR NA SILVER 
nome_tabela_silver = "olist.Silver.silver_products"
df_silver_products.write.format("delta") \
                .mode("overwrite") \
                .saveAsTable(nome_tabela_silver)
                
print(f"Tabela {nome_tabela_silver} criada com sucesso!")
```

```python
# PAYMENTS 
df_bronze_payments = spark.table("olist.Bronze.bronze_order_payments")

# 2. APLICAR TRANSFORMAÇÕES 

df_silver_payments = df_bronze_payments.withColumn(
    "payment_value", F.col("payment_value").cast("double")
) \
.withColumn(
    "payment_installments", F.col("payment_installments").cast("double")
)

# 3. SALVAR NA SILVER 
nome_tabela_silver = "olist.Silver.silver_payments"
df_silver_payments.write.format("delta") \
                .mode("overwrite") \
                .saveAsTable(nome_tabela_silver)
                
print(f"Tabela {nome_tabela_silver} criada com sucesso!")
```

```python
# --- Criando a fct_sales (Tabela Fato) ---

# 1. LER AS TABELAS SILVER
df_orders = spark.table("olist.Silver.silver_orders")
df_items = spark.table("olist.Silver.silver_order_items")
df_payments = spark.table("olist.Silver.silver_payments")

# 2. APLICAR JOINS PARA CRIAR A FATO
# O "grão" da nossa tabela fato será o "item de pedido"

# Join de Pedidos com Itens (Base)
df_fact = df_orders.join(
    df_items,
    "order_id",
    "inner"  # Queremos apenas pedidos que tenham itens
)

# Join com Pagamentos
# Cuidado: Um pedido pode ter múltiplos pagamentos. 
# Vamos agregar os pagamentos *antes* de fazer o join.
df_payments_agg = df_payments.groupBy("order_id") \
                             .agg(
                                 F.sum("payment_value").alias("total_payment_value"),
                                 F.max("payment_installments").alias("max_payment_installments"),
                                 F.first("payment_type").alias("primary_payment_type") # Simplificação
                             )

# Agora sim, juntamos a fato com os pagamentos agregados
df_fct_sales = df_fact.join(
    df_payments_agg,
    "order_id",
    "left" # Manter pedidos mesmo que não tenham pagamento (ex: cancelados)
)

# 3. SELECIONAR COLUNAS E SALVAR NA GOLD
# Vamos selecionar as "chaves" (FKs) e as "métricas" (fatos)
df_final_fact = df_fct_sales.select(
    # Chaves (para conectar com as Dimensões)
    "order_id",
    "product_id",
    "customer_id",
    "seller_id",
    
    # Métricas (Fatos)
    "price",
    "freight_value",
    "total_payment_value",
    "max_payment_installments",
    
    # Datas (importantes para análise de tempo)
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    
    # Outros atributos
    "order_status",
    "primary_payment_type"
)

# 4. SALVAR NA GOLD
nome_tabela_gold = "olist.Gold.fct_sales"
df_final_fact.write.format("delta") \
             .mode("overwrite") \
             .saveAsTable(nome_tabela_gold)

print(f"Tabela {nome_tabela_gold} criada com sucesso!")
```
